# Polynomial Regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Harish\Documents\MyML\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 6 - Polynomial Regression\Python\Position_Salaries.csv")

X= dataset.iloc[:,1:-1].values
Y = dataset.iloc[: ,-1].values
```

```python
from sklearn.linear_model import LinearRegression
li = LinearRegression()
li.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol = PolynomialFeatures(degree = 4)
X_poly = pol.fit_transform(X)
po_reg =LinearRegression()
po_reg.fit(X_poly,Y)
```

### With Linear Regression

```python
plt.scatter(X,Y)
plt.plot(X,li.predict(X) ,color ='red')
plt.title('Linear reg')
plt.xlabel('rank')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(X,po_reg.predict(X_poly) ,color ='red')
plt.title('poly reg')
plt.xlabel('rank')
plt.ylabel('salary')
plt.show()
```

### With Polynomial Regression

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color = 'red')
plt.plot(X_grid, po_reg.predict(pol.fit_transform(X_grid)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
